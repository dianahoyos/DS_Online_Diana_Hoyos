## Seleccin de Features (I): Motivacin y tipos

**MOTIVACION**

$$*$$

Por qu sera necesario seleccionar caractersticas? Para algunos, esta idea puede parecer contraintuitiva, ya que a ms datos en teora ms informacin.  

Al empezar el sprint ya vimos que exista lo que se denomina la maldicin de la "dimensionalidad" que puede a veces quedarse en algo ms etreo o difcil de ver. Vayamos a varias razones muy prcticas por las cuales merece la pena seleccionar y descartar o por lo menos priorizar tus features:

1. Un tema de **rendimiento** (a veces menor, con datasets pequeos): A menos features menos tiempos de procesado (en entrenamiento y en produccion)  

2. Un tema de ingenieria, **soporte y escalabilidad**: A menos features, los procesos de prepracin, ETL, y gestin son ms manejables. Adems a menos features, en general, menos probabilidad de fallos en esos procesos como consecuencia de los datos. (De nuevo depende muy mucho de los datasets).  

3. Un tema de **explicabilidad y atribucin** (cada vez ms necesario): A menos features son modelos ms fciles de atribuir (decidir que features o combinacin son las "importantes") y, en general, de explicar.[NOTA: Aqu mecanismos que transforman las features para luego reducirlas, i.e: PCA, puntan menos]  

4. Un tema de **robustez**: A menos features, menor probabilidad de que faltan datos para las mismas y el modelo no falle en produccin por falta de datos.  

5. Un tema de **regularizacion**: Menos features implican evitar problemas por los que algunos algoritmos confunden el ruido de las relaciones de algunas de estas y lo toman por informacin haciendo sobreajustes. Ya sabes, fastidia al modelo -> Lo haces ms fuerte.

Finalmente, hay un tema adicional de **conocimiento de los datos** que no tiene que ver con la seleccin sino con el proceso de "seleccionar", te obliga a conocer tus datos (a mayor o menor nivel dependiendo del tipo de "seleccin"). Este conocimiento siempre es bueno, ayuda a gestionar expectativas y a conocer donde atacar cuando surgen problemas o los resultados no son los esperados. 

Hay que reducir siempre las features? No. De hecho, muchas veces ni pasaras por este paso (errneamente). Siempre analiza tus features como si tuvieras que seleccionarlas. Toma notas de cuales son ms informativas, cuales estn ms relacionadas, cuales se podran transformar, por supuesto cuales son poco informativas, etc, etc. Priorizalas y luego salas todas si tu tiempo y tus recursos lo permiten, para luego hacer la seleccin. O no las uses todas y reservate las menos prioritarias, si los resultados acompaan, las tendrs ah para tiempos de necesidad.

Ahh, s, las features cuyo valor no cambia o es diferente para cada instancia (features invariantes y features indice) puedes obviarla para el modelo. Estas s, aunque ojo con los ndices para posteriores asignaciones e identificaciones.


**TIPOS y SUBTIPOS**

No es que existan formalmente unos tipos de seleccin de features, ms bien hay mecnicas que varan segn los datos de los que disponemos y de lo que queremos hacer con ellos y el nivel de profundidad del anlisis. En esa lnea distinguimos:

**A. Anlisis "visual"**: Que se apoya en  medios visuales y numricos (el que hemos empleado hasta ahora)  

**B. Anlisis con mtodos supervisados**: Se dan cuando tenemos un dataset con target y se apoya en relaciones numricas de las features con el target. Vamos a ver tres subtipos:  
    B.1 Filter: Seleccin mediante medidas matemticas de relacin (depende del tipo de target y del tipo de feature, categrica y numrica). Es la versin ms numrica del tipo anterior.  
    B.2 Intrinsic: Seleccin mediante un modelo. Se crea un modelo sencillo que prediga la variable target y se seleccionan las features de mayor importancia para trabajar en modelados posteriors ms complejos.  
    B.3 Wrapper: Seleccin mediante wrapper. Se crea un modelo con todas o con una nica feature, se hace una seleccin o deseleccin hasta llegar a un punto de compromiso.  

**C. Anlisis con mtodos no supervisados**: Nos vale para cualquier tipo de dataset y son mtodos basados en varianzas y covarianzas:  
    C.1 Umbral de Varianza: Se fija un umbral de varianza y se descartan aquellas variables que no lo alcanzan.  
    C.2 PCA: Comentada y usada en la unidad anterior espcialmente dedicada ella.

#### Mutual Information



Para obtener el coeficiente de Informacin Mutua (Mutual Information, MI) entre dos variables categricas en Python, usaremos la funcin `mutual_info_score` del mdulo `sklearn.metrics`. La Informacin Mutua es una medida de la cantidad de informacin que una variable contiene sobre otra. En el contexto de variables categricas, esto puede ayudarte a entender cunto se puede saber sobre una variable al observar la otra.

Es importante mencionar que la Informacin Mutua es simtrica: \(MI(X;Y) = MI(Y;X)\), lo que significa que el orden de las variables no afecta el resultado.

La Informacin Mutua es especialmente til en la seleccin de caractersticas para modelos de aprendizaje automtico, ya que permite identificar y seleccionar las caractersticas que comparten ms informacin con la variable objetivo, potencialmente mejorando el rendimiento del modelo.

```python
# Volviendo al titanic
from sklearn.metrics import mutual_info_score
features_cat = ["who","embarked","pclass","alone"]
target = "survived"
df_mi = df.dropna()
for col in features_cat:
    print(f"MI({col};{target}):", mutual_info_score(df_mi[col],df_mi[target]))
```

La informacin mutua se interpreta:

- Un valor de \(I(X; Y) = 0\) indica que \(X\) e \(Y\) son completamente independientes, es decir, el conocimiento de \(X\) no proporciona ninguna informacin sobre \(Y\), y viceversa.
- Un valor mayor de \(I(X; Y)\) indica una mayor dependencia o informacin compartida entre \(X\) y \(Y\).



#### Pearson y Spearman

Para obtener una correlacin entre dos series de datos en Python, podemos usa Pandas y NumPy, como ya sabes:

**Usando Pandas**

Si tus datos estn en estructuras de datos de Pandas (por ejemplo, `Series`), puedes utilizar el mtodo `corr` para calcular la correlacin. Pandas soporta varios mtodos de correlacin como Pearson (por defecto), Spearman y Kendall.


```python

# Crear dos series de datos
serie1 = pd.Series([1, 2, 3, 4, 5])
serie2 = pd.Series([5, 4, 3, 2, 1])

# Calcular la correlacin de Pearson
correlacion_pearson = serie1.corr(serie2)
print(f"Correlacin de Pearson: {correlacion_pearson}")

# Para calcular la correlacin de Spearman o Kendall, especifica el mtodo
correlacion_spearman = serie1.corr(serie2, method='spearman')
print(f"Correlacin de Spearman: {correlacion_spearman}")
```


**Usando NumPy**

NumPy no tiene una funcin directa para calcular correlaciones entre dos series, pero puedes calcular el coeficiente de correlacin de Pearson usando `numpy.corrcoef`, que devuelve la matriz de correlacin.


```python

import numpy as np

# Arrays de datos
datos1 = np.array([1, 2, 3, 4, 5])
datos2 = np.array([5, 4, 3, 2, 1])

# Calcular la matriz de correlacin
matriz_correlacion = np.corrcoef(datos1, datos2)

# El coeficiente de correlacin de Pearson est en la posicin [0, 1] y [1, 0]
print(f"Correlacin de Pearson: {matriz_correlacion[0, 1]}")
print(matriz_correlacion)

```

#### Categorica (no binaria y binaria)-Numerica: ANOVA

Lo empleamos para testar si existe relacin significativa entre una variable numrica y una categrica (cualquiera de ellas puede ser el target o simplemente por ver posible relacin entre dos features)

```python
from scipy import stats
```

```python
cat_col = "quality"
quality_cat = df_wines[cat_col].unique()  # Obtener los valores nicos de la columna categrica, en este caso la compaa rea
alcohol_por_quality = [df_wines[df_wines[cat_col] == categoria]['alcohol'] for categoria in quality_cat] # obtenemos los ingresos por compaa y los incluimos en una lista
```

```python
f_val, p_val = stats.f_oneway(*alcohol_por_quality) # El mtodo * (igual que cuando vimos *args hace mil aos) 
                                                    # lo que hace es separar todos los elementos de la lista y pasarselos como argumento a la funcin
print("Valor F:", f_val)
print("Valor p:", p_val)
```

Si el valor de p-value est por debajo de 1- Nivel de significacin deseado, entonces podemos aceptar la hiptesis alternativa de que existe relacin entre las variables a ese nivel de significacin.

Existe una forma de obtener las mejores k variables numricas en funcin de su **ANOVA con un target categrico**:

```python
from sklearn.feature_selection import SelectKBest, f_classif

'''
f_classif realiza un test ANOVA
a tcnica de anlisis de varianza (ANOVA) tambin conocida como anlisis factorial y desarrollada por Fisher en 1930,
constituye la herramienta bsica para el estudio del efecto de uno o ms factores (cada uno con dos o ms niveles) sobre
la media de una variable continua. Es por lo tanto el test estadstico a emplear cuando se desea comparar las medias de
dos o ms grupos.
'''
#features_num.remove("quality")
X_train = df_wines[features_num]
y_train = df_wines["class"]
selector = SelectKBest(f_classif, k=5)
x_data_kbest = selector.fit_transform(X_train, y_train)
X_train_kbest = pd.DataFrame(x_data_kbest, columns = selector.get_feature_names_out())
X_train_kbest

```

#### Categrica (binaria)- Numrica: U de Mann-Whitney

Cuando la categrica es binaria podemos hacer la prueba U para comprobar si existe una relacin estadsticamente significativa entre la features numrica y la categrica:

```python
from scipy.stats import mannwhitneyu
cat_col = "class"
num_col = "alcohol"

grupo_a = df_wines.loc[df_wines[cat_col] == "white"][num_col]
grupo_b = df_wines.loc[df_wines[cat_col] == "red"][num_col]

u_stat, p_valor = mannwhitneyu(grupo_a, grupo_b)

print("Estadstico U:", u_stat)
print("Valor p:", p_valor)
```

Podramos usar el valor de p como indicador de la fuerza de la relacin a p ms bajo menor seguridad de la existencia de relacin. En este caso el p-value de la prueba invita a considerar class como una feature predictora de alcohol como de segunda prioridad, por ejemplo.

#### Categrica-Categrica (Chi-2 ()):

Adems del Mutual Information introducido esta sesin, tenemos el test Chi-2 para evaluar si existe relacin entre dos features categricas:

```python
cat_1 = "pclass"
cat_2 = "who"

#Es necesario obtener la tabla de contingencia (sabemos hacerlo con Pandas):

tabla_contingencia = pd.crosstab(df[cat_1], df[cat_2])

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(tabla_contingencia)

print("Valor Chi-Cuadrado:", chi2)
print("P-Value:", p)
print("Grados de Libertad:", dof)
print("Tabla de Frecuencias Esperadas:\n", expected)
```

En este caso la relacin tiene una clara significacin estadstica.

#### B.3 Wrapper

Este tipo de mecanismos se basan en ir quitando o sumando features hasta alcanzar un umbral en una mtrica de seleccin (normalmente la mtrica que usaramos en el problema a resolver con el modelo que estemos buscando) o conseguir pocas variaciones por debajo o por encima del mismo o alcanzar un nmero de features determinado por el usuario (o sea nosotros).  

Veremos dos aproximaciones:
1. Partir de todas las features e ir entrenando modelos quitando cada vez la feature con menor peso hasta llegar a la condicin de parada.  
2. Partir de una feature e ir aadiendo features hasta alcanzar la condicin de parada  

#### Eliminacin recursiva de features

Se escoge un estimador o modelo y un numero n de features. Se lanza el estimador con TODAS
las features y se calcula el "feature importance" de estimador. La peor feature se
elimina y se tira otro modelo con una feature menos, asi hasta alcanzar n

Puedes verlo como una combinacin iterativa del selection by model en el que no haces un nico corte sino que vas haciendo cortes sucesivos para as captar los posibles cambios en las importancias de las features al tener cada vez menos.

En sklearn se implementa con `RFE` (Recursive Feature Elimination) y necesita de un modelo que tenga o bien el atributo `feature_importances_` o `coef_`

```python
from sklearn.feature_selection import RFE

X_train_wines = df_wines[features_num].copy()
y_train_wines = df_wines["class"]
```

```python
rf_RFE = RandomForestClassifier(random_state= 42) # Puedes usar tu modelo, o modelos a probar o un modelo "proxy" como en la seleccin por modelo

rfe = RFE(estimator = rf_RFE,
          n_features_to_select= 5, # Iterar hasta quedarse con 5
          step = 1) # Elimina de una en una, pero es configurable

rfe.fit(X_train_wines, y_train_wines)
print(rfe.ranking_)
pd.DataFrame(rfe.ranking_, columns= ["ranking"], index = features_num).sort_values("ranking")
```

#### Seleccin secuencial de Features

En este caso empezamos con una feature y vamos aadiendo hasta llegar a un lmite de features o bien hasta alcanzar un umbral o valor que no vara excesivamente (tendremos que tener una tolerancia).

En general funciona as:
1. Se entrena un modelo para cada feature con esa feature como predictora. Nos quedamos con la feature que consiga el mejor modelo (tendremos que seleccionar una mtrica)
2. Se entrena un modelo para cada combinacin de la feature escogida y el resto de features. Nos quedamos con la mejor combinacin.
3. Se repite 2 hasta llegar a la condicin de parada.

Por supuesto en "sklearn" tenemos una clase para hacerlo: `SequentialFeatureSelector`, que tiene la ventaja de que puede aplicarse aunque el modelo gua no tenga ni feature_imporances_ ni coef_ en sus atributos.

```python
from sklearn.feature_selection import SequentialFeatureSelector

rf_SFS = RandomForestClassifier(random_state = 42)

sfs_forward = SequentialFeatureSelector(rf_SFS,
                                        n_features_to_select = 5,
                                        cv = 4,
                                        scoring = "balanced_accuracy")


sfs_forward.fit(X_train_wines, y_train_wines)
```

Observa los tiempos de ejecucin, SFS es un mecanimos greedy (glotn) es decir que lo prueba todo, como el GridSearch, y por eso tardar tanto. Por otro lado, respecto a las elegidas aqu no hay ranking:

```python
print("Elegidasd SFS:", sfs_forward.get_feature_names_out())
```

Y de nuevo es diferente la seleccin. Hmmm, qu interesante.

#### Umbral de Varianza

Es el mecanismo ms sencillo y se basa en la idea de que las features que varan poco tienen poca informacin (al igual que las que son difererntes para todas las instancias)

Basado en lo anterior, escogemos un umbral de varianza mnima y seleccionamos aquellas features cuya varianza est por encima:

```python
from sklearn.feature_selection import VarianceThreshold

umbral_varianza = 0.5

var_selector = VarianceThreshold(umbral_varianza)
var_selector.fit(X_train_wines)
print("Lista seleccionada para selector de Varianza", var_selector.get_feature_names_out())
```

Y como todos los transformadores de `sklearn` permite obtener el array transformado es decir solo con las features seleccionadas:

```python
X_train_var = var_selector.transform(X_train_wines)
X_train_var
```

```python
X_train_var = pd.DataFrame(var_selector.transform(X_train_wines), columns= var_selector.get_feature_names_out())
X_train_var
```

Observa que no necesitamos un target para nada, con lo que es posible aplicarlo en cualquier dataset. Lo til o no que sea depende del problema y contexto.
