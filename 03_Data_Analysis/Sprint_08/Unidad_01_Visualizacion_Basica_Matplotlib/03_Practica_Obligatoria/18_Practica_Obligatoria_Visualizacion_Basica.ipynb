#1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

```python
# 1. Grfico de barras  Ciudades ms pobladas


import matplotlib.pyplot as plt
import numpy as np

# Seleccionar las 5 ciudades ms pobladas
top5 = df_clean.nlargest(5, "population_total")

colors = ["blue", "gray", "violet", "brown", "lightgreen"]

# crea la figura y idbujar el grafico de barras
plt.figure(figsize=(8,6))                    
bars = plt.bar(top5["city"], top5["population_total"], color=colors)

# Agregar etiquetas encima de cada barra
for b in bars:
    height = b.get_height()
    plt.text(b.get_x() + b.get_width()/2, height + 50,
             f"{height:,.0f}", ha="center", fontsize=12)

# Agregar el titulo y las etiquetas
plt.title("Ciudades ms pobladas de California", fontsize=15)
plt.ylabel("Habitantes (aprox.)")
plt.show()
```

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")

```python
# 2 Histograma  rea total ocupada por las ciudades

# crea un lienzo o espacio donde ir el grafico
plt.figure(figsize=(8,6))

# dibujar el histograma
plt.hist(df_clean["area_total_km2"],  # datos que queremos graficar
          bins=50, color="steelblue", # divide los datos en 50 grupos, mas bins,mas finas, steelblue color interno de la barras
            edgecolor="black", alpha=0.8)  # color del borde las barrras con el 0.8 que controla la transparencia
plt.title("Histograma del rea total ocupada por las ciudades", fontsize=15)  # Aadir titulo
plt.xlabel("Superficie (km)") # Etiquetas de los ejes
plt.ylabel("Nmero de Ciudades")
plt.legend(["Hist Area"])   # Agrega leyendas

plt.show()  # muestra el grafico
```

3. El scatter de rea de agua de un ciudad con su poblacion.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))

# --- Scatter general ---
plt.scatter(
    df_clean["area_water_km2"],
    df_clean["population_total"],
    color="green",
    alpha=0.6,
    s=35
)

plt.title("Poblacin por kilmetro cuadrado de agua", fontsize=14)
plt.xlabel("rea de agua en la ciudad ($km^2$)", fontsize=12)
plt.ylabel("Poblacin total aproximadamente", fontsize=12)

# --- Outliers segn tu grfico ---
out1 = (90, 3900)
out2 = (122, 1350)

# Dibujar puntos outliers
plt.scatter(*out1, color="green", s=10)
plt.scatter(*out2, color="green", s=30)

# Texto "Outliers"
px, py = 78, 3450
plt.text(px, py, "Outliers")

# Flechas
plt.annotate("", xy=out1, xytext=(px, py), arrowprops=dict(arrowstyle="->"))
plt.annotate("", xy=out2, xytext=(px, py), arrowprops=dict(arrowstyle="->"))


plt.show()

```

4. El boxplot de la altura de las ciudades.

```python
# 4. Boxplot  Altitud de las ciudades
plt.figure(figsize=(10,6))  # abrimos un lienzo para dibujar

# Creamos el boxplot
box = plt.boxplot(df_clean["elevation_m"],  # la columna que contiene la altitud de cada ciudad
                  vert=False,   # hace el boxplot horizontal
                  patch_artist=True,  # permite rellenar la caja con color
# Colores y estilos
                  boxprops=dict(facecolor="red", color="black"),
                  medianprops=dict(color="black"),
                  whiskerprops=dict(color="black"),
                  capprops=dict(color="black"),
                  flierprops=dict(marker='o', markersize=5))

plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.ylabel("Altitud (m)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("./data/california_cities.csv")
df_clean = df.dropna()   # quitar filas con nulos

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# 1) Barras: 5 ciudades ms pobladas
ax = axs[0, 0]

top5 = df_clean.nlargest(5, "population_total")

colors = ["blue", "gray", "violet", "brown", "lightgreen"]

bars = ax.bar(top5["city"], top5["population_total"], color=colors)

# Etiquetas numricas encima de cada barra
for b in bars:
    x = b.get_x() + b.get_width() / 2
    y = b.get_height()
    ax.text(x, y + 50, f"{y:.0f}", ha="center", va="bottom", fontsize=9)

ax.set_title("Ciudades ms pobladas de California", fontsize=13)
ax.set_ylabel("Habitantes (aprox.)", fontsize=11)
ax.tick_params(axis='x', rotation=45)

# 2) Histograma rea total ocupada
ax = axs[0, 1]

ax.hist(
    df_clean["area_total_km2"],
    bins=50,
    color="steelblue",
    edgecolor="black",
    alpha=0.8,
    label="Hist Area"
)

ax.set_title(
    "Histograma del rea total ocupada por las ciudades",
    fontsize=15,
    fontname="Bradley Hand ITC"  # si la fuente no existe, usa la por defecto
)
ax.set_xlabel("Superficie (km)", fontsize=11)
ax.set_ylabel("Nmero de Ciudades", fontsize=11)
ax.legend()

# 3) Scatter: rea de agua vs poblacin
ax = axs[1, 0]

ax.scatter(
    df_clean["area_water_km2"],
    df_clean["population_total"],
    color="green",
    alpha=0.6,
    s=35
)

ax.set_title("Poblacin por kilmetro cuadrado de agua", fontsize=13)
ax.set_xlabel("rea de agua en la ciudad ($km^2$)", fontsize=11)
ax.set_ylabel("Poblacin total aproximadamente", fontsize=11)

# Outliers (aproximados segn la figura de ejemplo)
out1 = (90, 3900)
out2 = (122, 1350)

ax.scatter(*out1, color="green", s=50)
ax.scatter(*out2, color="green", s=50)

# Texto "Outliers"
px, py = 75, 3450
ax.text(px, py, "Outliers")

# Flechas hacia los outliers
ax.annotate("", xy=out1, xytext=(px, py),
            arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=out2, xytext=(px, py),
            arrowprops=dict(arrowstyle="->"))

# 4) Boxplot horizontal de la altitud
ax = axs[1, 1]

ax.boxplot(
    df_clean["elevation_m"],
    vert=False,           # horizontal
    patch_artist=True,
    boxprops=dict(facecolor="red", color="black"),
    medianprops=dict(color="black"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    flierprops=dict(marker='o', markerfacecolor='black', markersize=5)
)

ax.set_title("Diagrama de caja de la Altitud de las ciudades de California")
ax.set_xlabel("Altitud (m)")

# Ajustar espacios para que no se solapen
plt.tight_layout()
plt.savefig("figura_ejercicio.png", dpi=300, bbox_inches="tight")
plt.show()
```
